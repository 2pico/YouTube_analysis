```python
# pull in the raw data
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df_ca = pd.read_csv('youtube-new/CAvideos.csv')
df_de = pd.read_csv('youtube-new/DEvideos.csv')
df_fr = pd.read_csv('youtube-new/FRvideos.csv')
df_gb = pd.read_csv('youtube-new/GBvideos.csv')
df_in = pd.read_csv('youtube-new/INvideos.csv')
#df_jp = pd.read_csv('youtube-new/JPvideos.csv')
#df_kr = pd.read_csv('youtube-new/KRvideos.csv')
#df_mx = pd.read_csv('youtube-new/MXvideos.csv')
#df_ru = pd.read_csv('youtube-new/RUvideos.csv')
df_us = pd.read_csv('youtube-new/USvideos.csv')

df_us.info()
```

```python
# add the country code to the tables before merging them
df_ca['country'] = 'ca'
df_de['country'] = 'de'
df_fr['country'] = 'fr'
df_gb['country'] = 'gb'
df_in['country'] = 'in'
df_us['country'] = 'us'

df_us['country'].head()
```

```python
# concatinate the dataframes into one giant dataframe
frames = [df_ca, df_de, df_fr, df_gb, df_in, df_us]

df_master = pd.concat(frames)
df_master.info()
```
